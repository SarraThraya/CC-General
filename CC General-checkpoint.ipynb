```python
import pandas as pd
df=pd.read_csv (r'C:\Users\mes documents\Downloads\CC GENERAL.csv', sep=',')
df
```

```python
df.drop('CUST_ID',axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df['MINIMUM_PAYMENTS'].fillna(df['MINIMUM_PAYMENTS'].mean(), inplace=True)
```

```python
df['CREDIT_LIMIT'].fillna(df['CREDIT_LIMIT'].mean(), inplace=True)
```

```python
from sklearn.cluster import AgglomerativeClustering #Importing our clustering algorithm : Aggl
import numpy as np
model=AgglomerativeClustering(n_clusters=18, affinity='euclidean', linkage='complete')
clust_labels=model.fit_predict(df)  #Applying agglomerative algorithm with 18 clusters, using euclidean distance as a metric
```

```python
agglomerative=pd.DataFrame(clust_labels)
agglomerative
```

```python
import matplotlib.pyplot as plt
fig =plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter (df ["PURCHASES"] , df ["PURCHASES_FREQUENCY"] , c= agglomerative[0], s=50)
ax.set_title("Agglomerative Clustering")
ax.set_xlabel("PURCHASES")
ax.set_ylabel("PURCHASES_FREQUENCY")
plt.colorbar(scatter)
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("Customer Dendrograms")
dend=shc.dendrogram(shc.linkage(df, method="complete"))
```

```python
from sklearn.cluster import KMeans  #Importing our clustering algorithm: KMeans
kmeans=KMeans(n_clusters=5, random_state=0)  #Cluster our data by choosing 5 as number of clusters
kmeans.fit(df)
```

```python
labels=pd.DataFrame(kmeans.labels_)
labels
```

```python
kmeans.predict(df)
print(kmeans.cluster_centers_)   #Printing the coordinates of cluster centers.
```

```python
df['labels']=labels
plt.scatter(df["PURCHASES"][df.labels == 0],df["PURCHASES_FREQUENCY"][df.labels == 0],s=80,c='magenta',label='Careful')
plt.scatter(df["PURCHASES"][df.labels == 1],df["PURCHASES_FREQUENCY"][df.labels == 1],s=80,c='yellow',label='Standard')
plt.scatter(df["PURCHASES"][df.labels == 2],df["PURCHASES_FREQUENCY"][df.labels == 2],s=80,c='green',label='Target')
plt.scatter(df["PURCHASES"][df.labels == 3],df["PURCHASES_FREQUENCY"][df.labels == 3],s=80,c='cyan',label='Careless')
plt.scatter(df["PURCHASES"][df.labels == 4],df["PURCHASES_FREQUENCY"][df.labels == 4],s=80,c='burlywood',label='Sensible')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label = 'Centroids')
plt.title('Clusters of Customers')
plt.xlabel('PURCHASES')
plt.ylabel('PURCHASES_FREQUENCY')
plt.legend()
plt.show()
```

```python
Sum_of_squared_distances = []
k =range(1,15)
for k in k:
    km=KMeans(n_clusters=k)
    km=km.fit(df)
    Sum_of_squared_distances.append(km.inertia_)
```
